```python
print('Class', df.Class.describe())
print()
print('Class 0: ', df[df.Class ==0].Class.count())
print('Class 1: ', df[df.Class ==1].Class.count())
```

```python
###class가 기준으로 시간 단위로 bar chart

f, (ax1, ax2) = plt.subplots(2,1, sharex = True, figsize =(12,4))

ax1.hist(df.Time[df.Class == 1], bins = 50)
ax1.set_title('Ones')

ax2.hist(df.Time[df.Class == 0], bins = 50)
ax2.set_title('Zeros')

plt.xlabel('Time')
plt.ylabel('Count')
plt.show()
```

```python
### amount단위로

f, (ax1, ax2) = plt.subplots(2,1, sharex = True, figsize =(12,4))

ax1.hist(df.Amount[df.Class == 1], bins = 30)
ax1.set_title('Ones')

ax2.hist(df.Amount[df.Class == 0], bins = 30)
ax2.set_title('Zeros')

plt.xlabel('Amount')
plt.ylabel('N??')
plt.yscale('log')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2,1, sharex = True, figsize =(12,6))

ax1.scatter(df.Time[df.Class == 1], df.Amount[df.Class == 1])
ax1.set_title('Ones')

ax2.scatter(df.Time[df.Class == 0], df.Amount[df.Class == 0], color = 'red')
ax2.set_title('Zeros')

plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```python
plt.scatter(df.Time[df.Class == 0], df.Amount[df.Class == 0], c = 'b')
plt.scatter(df.Time[df.Class == 1], df.Amount[df.Class == 1], c = 'r')


plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```python
v_features = df.iloc[:, 1:29].columns
```

```python
plt.figure(figsize = (12, 28*4))
aa = plt.GridSpec(29,1)
for i, cn in enumerate(df[v_features]):
    ax = plt.subplot(aa[i])
    sns.distplot(df[cn][df.Class == 0], bins = 50, color = 'red')
    sns.distplot(df[cn][df.Class == 1], bins = 50, color = 'green')
    ax.set_title('histogram of feature: ' + str(cn))

plt.show()
```
